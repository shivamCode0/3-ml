```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use(['science', 'no-latex'])

# Change size back to default
plt.rcParams.update({'figure.figsize': (5, 3), 'figure.dpi': 300})

# Data from the provided chart
labels = ['Wave 1\n(to early 2020s)', 'Wave 2\n(to late 2020s)', 'Wave 3\n(to mid-2030s)']
men = [2.5, 17.0, 30.0]
women = [3.5, 18.5, 32.0]
all_ = [2.5, 17.5, 30.0]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()

rects1 = ax.barh(x - width, men, width, label='Men', color='#E69F00')
rects2 = ax.barh(x, all_, width, label='All', color='#000000')
rects3 = ax.barh(x + width, women, width, label='Women', color='#D55E00')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('% of existing jobs at potential risk of automation')
ax.set_title('% of existing jobs at potential risk of automation')
ax.set_yticks(x)
ax.set_yticklabels(labels)
ax.legend()


# Add source text
plt.figtext(0.5, -0.1, 'Source: PwC estimates based on OECD PIAAC data (median values for 29 countries)', ha='center', fontsize=8)

fig.tight_layout()

plt.show()
```
